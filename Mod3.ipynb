## shallow and deep copy

```python
# shallow copy, creates a new variable that shares the reference of the original object. Any changes made to a copy of object will be reflected in the original object.
sample = cars_data.copy(deep=False)
sample.head(20)
```

```python
# deep copy, creates a new variable that stores the values of the original object. Any changes made to the original object will not be reflected in the new object
sample = cars_data.copy(deep=True)
sample.head(20)
```
