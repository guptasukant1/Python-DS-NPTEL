## How to fill missing values

With mean/median for numerical variable\
With class which has maximum count for categorical variable

```python
missing
```

Median Value is prefered as if the values are biased towards lower end or the higher end, then the missing values will tip toward that specifc side.\
This can lead to misleading data

```python
df2.describe()
# $ Helps to show the descriptive stats of the dataFrame, summarize central tendency, dispersion and shape of a dataset’s distribution, excluding NaN values.
```

As we can see that the mean and 50%Q values are similar for the Age, KM, and HP, thus we can select mean as our measure for missing values of Age, median for KM, and mean for HP\
For FuelType, we can find the mode of the types of fuels, and for MetColor we can find the mode and use the 0th index of the same

```python
df2['Age'].mean()
```

```python
df2['Age'].fillna(df2['Age'].mean(), inplace = True)
```

```python
df2.isnull().sum()
```

```python
df2['KM'].median()
```

```python
df2['KM'].fillna(df2['KM'].median(), inplace=True)
```

```python
df2.isnull().sum()
```

```python
df2['HP'].mean()
```

```python
df2['HP'].fillna(df2['HP'].mean(), inplace= True)
```

```python
df2.isnull().sum()
```

```python
print(df2['FuelType'].value_counts())
print(df2['FuelType'].value_counts().index[0])
```

```python
df2['FuelType'].fillna(df2['FuelType'].value_counts().index[0], inplace=True) #type: ignore
```

```python
df2.isnull().sum()
```

```python
df2['MetColor'].mode()
```

```python
df2['MetColor'].fillna(df2['MetColor'].mode().index[0], inplace = True) #type:ignore
```

```python
df2.isnull().sum()
```

### Entering values using lambda function to deal with multiple null attributes at once for both classification and regression problems

```python
df4 = df.copy(deep = True)
df4
```

```python
df4.isnull().sum()
```

```python
df4 = df4.apply(lambda x: x.fillna(x.mean())
                if x.dtype == 'float'
                else
                x.fillna(x.value_counts().index[0]))
```

```python
df4.isnull().sum()
```

```python
df4
```

```python
df4['Doors'].replace("three", 3, inplace = True)
df4['Doors'].replace("four", 4, inplace = True)
df4['Doors'].replace("five", 5, inplace = True)
```

```python
df4['Doors'] = df4['Doors'].astype('int64')
# df4
```
