```python
def c_convert(x):
    val_convert = x / 12
    return val_convert

cars_data1['Age_Converted'] = c_convert(cars_data1['Age'])
cars_data1['Age_Converted'] = round(cars_data1['Age_Converted'], 1)
```

```python
cars_data1.insert(12, 'KM_per_month', 0)
```

```python
def KM_convert(x1, x2):
    val_convert = x1/12
    ratio = x2/x1
    return [val_convert, ratio]

cars_data1['Age_Converted'], cars_data1['KM_per_month'] = KM_convert(cars_data1['Age'], cars_data1['KM'])
cars_data1['Age_Converted'] = round(cars_data1['Age_Converted'], 1)
```

# Extrapolatory Data Analysis

```python
pd.crosstab(index = cars_data2['FuelType'], columns = 'count', dropna = True) # type: ignore
# $ This is a frequency table for the FuelType column
```

```python
pd.crosstab(index = cars_data2['Automatic'], columns = cars_data2['FuelType'], dropna = True)
# $ Two way tables
```

```python
pd.crosstab(index = cars_data2['Automatic'], columns = cars_data2['FuelType'], normalize = True, dropna = True)
# $ 2 way table - joint probability [likelihood of 2 independent events happening together]
# $ normalize = True gives the joint probability (converts all the values in the table to proportion)
```

```python
pd.crosstab(index = cars_data2['Automatic'], columns = cars_data2['FuelType'], margins = True, normalize = True, dropna = True)
# $  way table - marginal probability [likelihood of a single event happening]
# $ margins = True gives the marginal probability (row and column totals are added to the table)
# $ All column and All row are part of marginal probability
```

```python
pd.crosstab(index = cars_data2['Automatic'], columns = cars_data2['FuelType'], margins = True, dropna = True, normalize = 'index')
# $ 2 way table- Conditional probability
# $ Conditional probability is the likelihood of an event (A) given that another event (B) has already occurred
```

```python
pd.crosstab(index = cars_data2['Automatic'], columns = cars_data2['FuelType'], margins = True, dropna = True, normalize = 'columns')
# $ 2 way table- Conditional probability
# $ Conditional probability is the likelihood of an event (A) given that another event (B) has already occurred
```

# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
cars_data.dropna(axis=0, inplace=True)
```

### Scatter Plot
set of points that represent the values for 2 diff variables plotted on x and y axes

Used to convey the relationship b/w 2 num vars

correlation plots

```python
plt.scatter(cars_data['Age'], cars_data['Price'], c='red')
plt.title('Scatter plot of Price vs Age of the cars')
plt.xlabel('Age (months)')
plt.ylabel('Price (Euros)')
plt.show()
```

### Histogram
graphical representation of data using bars of diff heights

For frequncy distribution of num vals

```python
plt.hist(cars_data['KM'])
plt.hist(cars_data['KM'],
         color='green', edgecolor='white', bins=5)
plt.title('Histogram of Kilometer')
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
plt.show()
```

### Bar plot
shows categorical with rectangular bars of diff lengths

For freq distribution of categorical vars

easy to compare sets of data b/w diff groups

```python
counts = [979, 120, 12]
fuelType = ('Petrol', 'Diesel', 'CNG')
index = np.arange(len(fuelType))

plt.bar(index, counts, color = ['red', 'blue', 'cyan'])
plt.title('Bar plot of Fuel Type')
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')
plt.xticks(index, fuelType, rotation = 90)
plt.show()
```

# Data Visualization using seaborn

```python
sns.set_theme(style = 'darkgrid')
sns.regplot(x = cars_data['Age'], y = cars_data['Price'], marker = "*", fit_reg = False)
```

### Linear Model Plot

Combination of scatter plot and linear regression fit

```python
sns.lmplot(
    x = 'Age',
    y = 'Price',
    data = cars_data,
    fit_reg = False,
    hue = 'FuelType',
    legend = True,
    palette = 'Set1')
# $ hue is used to color the data points based on the categories in the FuelType
```

### Box and whiskers plot - numerical variable
It is a summary of the dataset. Has 5 key stats: minimum, 1st Auartile, 2nd Quartile(median), 3rd Quartile, and maximum.

box: b/w the 1st and 3rd Q and show spread of the 50% of data\
whiskers: extend from box and show range of data.\
Outliers: data points beyond range of whiskers.

```python
sns.boxplot(y = cars_data['Price'])
```

### Box and whiskers plot for numerical vs categorical variables along with grouped B&W plot
